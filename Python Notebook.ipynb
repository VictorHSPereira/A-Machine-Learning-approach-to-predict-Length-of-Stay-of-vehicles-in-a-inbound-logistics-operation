## **Importando base de dados de BRM**
Base de dados de Janeiro 2021



```python
import pandas as pd
import numpy as np
import time

uri = "https://gist.githubusercontent.com/VictorHSPereira/c6fcf91dc133c1e003d1492af173424b/raw/1e4f459ecc6c13a86da5323a401d04cc785ec63c/ds_janeiro.CSV"
dados = pd.read_csv(uri, sep=';',decimal=',')
dados.dropna(inplace=True)
    
dados.head(5)
```

```python
x=dados[["dia_semana","Quinzena","agrup_veiculo","int_tipo_expedicao","bin_varricao","tempo_varricao","chave_ponto_descarga","chave_materiais","multimateriais","peso_kg","qtd_veiculos_dia","qtd_veiculos_dentro"]]
y=dados["tpv_usina"]
```

### Linear Model Ridge

```python
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import max_error
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np

#Separando as bases de treino e teste
from sklearn.model_selection import train_test_split

Resultados = pd.DataFrame(columns = ['RMSE','NDEI','MAE', 'Time (ms)'])

SEED = 1

while SEED <= 50 :
  inicio = time.time()

  treino_x, teste_x, treino_y, teste_y = train_test_split(x,y,random_state = SEED, test_size = 0.15)


  modelo = linear_model.Ridge(alpha=0.1)
  modelo.fit(treino_x,treino_y)
  previsoes = modelo.predict(teste_x)

  MAE = mean_absolute_error(teste_y, previsoes)
  R2 = r2_score(teste_y, previsoes)
  RMSE = mean_squared_error(teste_y, previsoes)
  NDEI = RMSE/np.std(teste_y)

  # print("Linear Model Ridge")
  # print("Mean Absolute Error (MAE): %.2f" % MAE)
  # print("RMSE: %.2f" % RMSE)
  # print("Coeficiente de determinação (R2): %.2f" % R2)
  # print("NDEI: %.2f" % NDEI)

  # plt.plot(teste_y.values[-100:], label='Valor real')
  # plt.plot(previsoes[-100:], label='Previsao')
  # plt.legend()
  # plt.ylabel('TPV')
  # plt.show()

  fim = time.time()
  duration = fim - inicio

  Resultados=Resultados.append({'RMSE' : RMSE , 'NDEI' : NDEI, 'MAE' : MAE, 'Time (ms)': duration} , ignore_index=True)
  
  SEED = SEED + 1

Resultados.to_excel("Linear Model Ridge - JAN.xlsx")

print("RMSE mean:", Resultados['RMSE'].mean())
print("RMSE std:", Resultados['RMSE'].std())
print("NDEI mean:", Resultados['NDEI'].mean())
print("NDEI std:", Resultados['NDEI'].std())
print("MAE mean:", Resultados['MAE'].mean())
print("MAE std:", Resultados['MAE'].std())
print("Time (ms) mean:", Resultados['Time (ms)'].mean())
print("Time (ms) std:", Resultados['Time (ms)'].std())
```

### KNeighborsRegressor

### GradientBoostingRegressor

```python
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import max_error
from sklearn.metrics import mean_squared_error
import numpy as np

#Separando as bases de treino e teste
from sklearn.model_selection import train_test_split

Resultados = pd.DataFrame(columns = ['RMSE','NDEI','MAE', 'Time (ms)'])

SEED = 1

params = {'n_estimators': 500,
            'max_depth': 4,
            'min_samples_split': 5,
            'learning_rate': 0.15,
            'loss': 'squared_error'}

while SEED <= 50 :
  inicio = time.time()

  treino_x, teste_x, treino_y, teste_y = train_test_split(x,y,random_state = SEED, test_size = 0.15)

  modelo = ensemble.GradientBoostingRegressor(**params)
  modelo.fit(treino_x,treino_y)
  previsoes = modelo.predict(teste_x)

  MAE = mean_absolute_error(teste_y, previsoes)
  R2 = r2_score(teste_y, previsoes)
  # max_error = max_error(teste_y, previsoes)
  RMSE = mean_squared_error(teste_y, previsoes)
  NDEI = RMSE/np.std(teste_y)

  # print("GradientBoostingRegressor")
  # print("Mean Absolute Error (MAE): %.2f" % MAE)
  # print("RMSE: %.2f" % RMSE)
  # print("Coeficiente de determinação (R2): %.2f" % R2)
  # print("Erro máximo: %.2f" % max_error)
  # print("NDEI: %.2f" % NDEI)

  # import matplotlib.pyplot as plt
  # plt.plot(teste_y.values[-373:], label='Valor real')
  # plt.plot(previsoes[-373:], label='Previsao')
  # plt.legend()
  # plt.ylabel('TPV')
  # plt.show()
  
  fim = time.time()
  duration = fim - inicio

  Resultados=Resultados.append({'RMSE' : RMSE , 'NDEI' : NDEI, 'MAE' : MAE, 'Time (ms)': duration} , ignore_index=True)
  
  SEED = SEED + 1

Resultados.to_excel("GradientBoostingRegressor - JAN.xlsx")  

print("RMSE mean:", Resultados['RMSE'].mean())
print("RMSE std:", Resultados['RMSE'].std())
print("NDEI mean:", Resultados['NDEI'].mean())
print("NDEI std:", Resultados['NDEI'].std())
print("MAE mean:", Resultados['MAE'].mean())
print("MAE std:", Resultados['MAE'].std())
print("Time (ms) mean:", Resultados['Time (ms)'].mean())
print("Time (ms) std:", Resultados['Time (ms)'].std())


```
